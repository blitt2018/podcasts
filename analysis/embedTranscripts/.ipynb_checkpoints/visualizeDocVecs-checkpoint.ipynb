```python
import gensim
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import numpy as np
import pandas as pd
from sklearn.manifold import TSNE
from sklearn.decomposition import TruncatedSVD, PCA
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
MODEL_PATH = "/shared/3/projects/benlitterer/podcastData/embeddings/floydMonth/doc2vecModel5000"
model = Doc2Vec.load(MODEL_PATH)
```

```python
embeddings = np.array([model.dv[i] for i in range(5000)])
```

```python
df = pd.read_csv("/shared/3/projects/benlitterer/podcastData/processed/floydMonth/floydMonthEnSHORT.csv", nrows=5000)
```

```python
tsne = TSNE(perplexity=150)
```

```python
tsneDims = tsne.fit_transform(embeddings)
```

```python
df[["tsne1", "tsne2"]] = tsneDims
dfTsne = df.dropna(subset=["category1"])
```

```python
sns.scatterplot(data=dfTsne[dfTsne["category1"].str.contains("religion|business")], x="tsne1", y="tsne2", hue="category1")
```

```python
svd = TruncatedSVD()
svdEmbeddings = svd.fit_transform(embeddings)
df[["svd1", "svd2"]] = svdEmbeddings
```

```python
dfSVD = df.dropna(subset=["category1"])  
sns.scatterplot(data=dfSVD[dfSVD["category1"].str.contains("religion|business")], x="svd1", y="svd2", hue="category1", alpha=.2)
```

```python
svd.explained_variance_ratio_
```

```python
pca = PCA(n_components=2)
pcaEmbeddings = pca.fit_transform(embeddings)
df[["pca1", "pca2"]] = pcaEmbeddings
```

```python
dfPCA = df.dropna(subset=["category1"])  
sns.scatterplot(data=dfPCA[dfPCA["category1"].str.contains("religion|business")], x="pca1", y="pca2", hue="category1", alpha=.2)
```
