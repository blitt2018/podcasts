```python
import pandas as pd
```

```python
transcriptDf = pd.read_json("/shared/3/projects/benlitterer/podcastData/processed/floydMonth/floydMonthDataClean.jsonl", orient="records", lines=True)
```

```python
speakerCountDf = pd.read_json("/shared/3/projects/benlitterer/podcastData/detectHostNumber/hostNumDf.jsonl", orient="records", lines=True)
```

```python
transcriptDf["diarizeOutPath"] = "/shared/3/projects/benlitterer/podcastData/diarization/mayJune" + transcriptDf["potentialOutPath"].apply(lambda x: x.replace("MERGED", ".rttm"))
```

```python
transcriptDf = transcriptDf.set_index("diarizeOutPath")
```

```python
pathIndices = list(speakerCountDf["path"])
```

```python
pathIndices[0]
```

```python
len(pathIndices)
```

```python
transcriptDf.head() 
```

```python
#we can only get transcripts we've actually transcribed!
commonIndices = list(set(transcriptDf.index).intersection(pathIndices))
len(commonIndices)
```

```python
transDiarized = transcriptDf.loc[commonIndices, :].reset_index()
```

```python
dups = transDiarized.duplicated(subset=["diarizeOutPath"], keep=False)
transDiarized = transDiarized.drop_duplicates(subset=["potentialOutPath"])
```

```python
speakerCountDf.head() 
```

```python
outDf = pd.merge(transDiarized, speakerCountDf.rename(columns={"path":"diarizeOutPath"}), on="diarizeOutPath", how="inner")
```

```python
outDf.to_json("/shared/3/projects/benlitterer/podcastData/detectHostNumber/mergedHostNumData.jsonl", orient="records", lines=True)
```
