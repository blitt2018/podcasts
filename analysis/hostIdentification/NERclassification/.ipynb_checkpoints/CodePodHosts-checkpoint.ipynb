```python
import pandas as pd
from fuzzywuzzy import fuzz
import matplotlib.pyplot as plt
```

```python
df = pd.read_json("/shared/3/projects/benlitterer/podcastData/hostIdentification/logisticRegression/allEntityFeatures.jsonl", lines=True, orient="records")
```

```python
def getLevenDist(inRow):
    return fuzz.ratio(inRow["ent"], inRow["itunesAuthor"]) 
```

```python
df = df.head(50000) 
```

```python
df["levenDist"] = df.apply(getLevenDist, axis=1)
```

```python
plt.hist(df["levenDist"]) 
plt.title("Levenshtein Distance - entity and itunesAuthor") 
```

```python

df["entLen"] = df["ent"].apply(lambda x: len(str(x).split()))
df["authLen"] = df["itunesAuthor"].apply(lambda x: len(str(x).split()))
```

```python
#code correct and incorrect
currSample = df.sample(500, random_state=25)
```

```python
df["twoWordEnt"] = df["ent"].apply(lambda x: len(x.split()) == 2)
df["twoWordAuth"] = df["itunesAuthor"].apply(lambda x: len(x.split()) == 2) 
```

```python
#get some positive examples 
posSample = df[(df["itunesMatch"] == True) & (df["twoWordEnt"] == True) & (df["twoWordAuth"] == True)].sample(500, random_state=25).reset_index(drop=True) 
posSample.to_csv("/shared/3/projects/benlitterer/podcastData/hostIdentification/logisticRegression/500pos.csv")
```

```python
negSample["levenDist"] = negSample.apply(getLevenDist, axis=1) 
```
