```python
import pandas as pd 
import csv
```

```python
df = pd.read_csv("/shared/3/projects/benlitterer/podcastData/processed/mergedMetaDataClean.csv")
```

```python
df["cleanDatesLoc"] = pd.to_datetime(df["cleanDatesLoc"])
```

```python
firstCutoff = pd.to_datetime("05-04-2020", format="%m-%d-%Y", utc=True)
lastCutoff = pd.to_datetime("05-10-2020", format="%m-%d-%Y", utc=True)

beforeFloydMonth = df[(df["cleanDatesLoc"] >= firstCutoff) & (df["cleanDatesLoc"] <= lastCutoff)]
beforeFloydMonth.to_json("/shared/3/projects/benlitterer/podcastData/processed/beforeFloydMonth/5_4_5_10.jsonl", orient="records", lines=True)
```

```python
print(f"5-4-20 through 5-10-20: {beforeFloydMonth.shape}") 
```
