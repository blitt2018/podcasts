```python
import os
import time

import numpy as np
import pandas as pd
import sys 

import audb
import audiofile
import opensmile
```

```python
#IN_PATH = sys.argv[1]
IN_PATH = "/shared/3/projects/benlitterer/podcastData/mp3s/testRound/11678693-episode-28-about-abusers-examining-change-part-1.wav"
#OUT_PATH = sys.argv[2]
OUT_PATH = "/shared/3/projects/benlitterer/podcastData/transcripts/premadeTransTranscripts/11678693-episode-28-about-abusers-examining-change-part-1.wav" 
```

```python
signal, sampling_rate = audiofile.read(
    IN_PATH,
    always_2d=True,
)
```

```python
#note increasing num_workers doesn't seem to help much here 
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.eGeMAPSv02,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors, 
    num_workers=2, 
    multiprocessing=True
)
```

```python
oldTime = time.time()
df = smile.process_signal(
    signal,
    sampling_rate
)
df.to_csv(OUT_PATH.split(".")[0] + "LowLevel.csv")
print(f"{time.time() - oldTime}")
```
