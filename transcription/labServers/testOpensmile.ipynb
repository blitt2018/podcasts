```python
import os
import time

import numpy as np
import pandas as pd

import audb
import audiofile
import opensmile

import seaborn as sns

```

```python
TEST_PATH = "/shared/3/projects/benlitterer/podcastData/mp3s/testRound/11678693-episode-28-about-abusers-examining-change-part-1.wav"
```

```python
signal, sampling_rate = audiofile.read(
    TEST_PATH,
    duration=60,
    always_2d=True,
)
```

```python
#note increasing num_workers doesn't seem to help much here 
smile = opensmile.Smile(
    feature_set=opensmile.FeatureSet.GeMAPSv01b,
    feature_level=opensmile.FeatureLevel.LowLevelDescriptors
)
smile.feature_names
```

```python
oldTime = time.time()
df = smile.process_signal(
    signal,
    sampling_rate
)
print(f"{time.time() - oldTime}")
```
