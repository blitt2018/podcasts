```python
def getPairs(inList): 
    inList = list(set(inList))
    outList = []
    for i in range(len(inList)):
        if i + 1 < len(inList): 
            for j in range(i+1, len(inList)): 
                outList.append([inList[i], inList[j]])
    return outList
```

```python
epDf = pd.read_csv("/shared/3/projects/benlitterer/podcastData/analysis/allEpisodes.csv", index_col=0, lineterminator='\n')
```

```python
entDf = entDf.merge(epDf[["key", "title"]], on="key", how="left")
```

```python
podDf = entDf[["title", "namedEnt", "key"]].groupby("title").agg(list)
```

```python
#get overlap from two lists of entities 
def getEntOverlap(l1, l2): 
    l1 = set(l1)
    l2 = set(l2)
    
    return (list(l1.intersection(l2)), len(l1.union(l2)))
```

```python
#get an inverted showing entities associated with podcasts 
#we don't care about multiple episodes for now, just if an entity is mentioned by a podcast 
entPodDf = entDf[["namedEnt", "title"]].drop_duplicates().groupby("namedEnt").agg(list)
entPodDf["numEnts"] = entPodDf["title"].apply(len)
entPodDf = entPodDf[entPodDf["numEnts"] > 1]
```

```python
entPodDf = entPodDf.sort_values("numEnts")
```

```python
#punch in here 
entPodDf.shape
```

```python
titleLists = entPodDf["title"].head(800000)

simTups = []
#go through all lists of lists 
for titleList in tqdm(titleLists): 
    
    for i in range(len(titleList)): 
        for j in range(i, len(titleList)): 
            p1 = titleList[i]
            p2 = titleList[j]
            
            if p1 != p2: 
                try: 
                    l1 = podDf.loc[p1, "namedEnt"]
                    l2 = podDf.loc[p2, "namedEnt"]

                    simEnts, union = getEntOverlap(l1, l2)
                    
                except: 
                    simEnts = []
                    union = []
            
                simTups.append([p1, p2, simEnts, union])
```

```python
podSims = pd.DataFrame(simTups, columns=["p1", "p2", "commonPods", "unionSize"]).dropna().drop_duplicates(subset=["p1", "p2"])
```

```python
p1u = set(podSims["p1"].unique())
p2u = set(podSims["p2"].unique())
allSimPods = list(p1u.union(p2u))
len(allSimPods)
```

```python
podSims["simCount"] = podSims["commonPods"].apply(len)
podSims["propCommon"] = podSims["simCount"] / podSims["unionSize"]
podSims = podSims.sort_values("propCommon").reset_index(drop=True)
```

```python
podSimsF = podSims[podSims["propCommon"] > .08]
```

```python
#we only capture about 1/10 podcasts this way 
p1u = set(podSimsF["p1"].unique())
p2u = set(podSimsF["p2"].unique())
len(p1u.union(p2u))
```

```python
myGraph = nx.from_pandas_edgelist(podSimsF, target="p1", source="p2", edge_attr="propCommon")
```

```python
comms = nx.community.louvain_communities(myGraph, seed=123, weight=.01)
comps = list(nx.connected_components(myGraph))
```
