```python
import pandas as pd
import os
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
scoreList = []
annotatorList = []
stem = "/shared/3/projects/benlitterer/podcastData/annotation/label1000/MACE/outputs"
for fPath in os.listdir(stem): 
    if "competence" in fPath: 
        currFile = open(stem + "/" + fPath)
        currScores = currFile.readlines()[0].strip("\n").split("\t")
        annotatorList += fPath.split(".")[0].split(",")
        scoreList += currScores
```

```python
scoreDf = pd.DataFrame({"annotators":annotatorList, "scores":scoreList})
scoreDf["scores"] = scoreDf["scores"].astype(float)
```

```python
fig, ax = plt.subplots(figsize=(3, 5))
lowScoreDf = scoreDf.sort_values("scores").head(10)
sns.barplot(lowScoreDf, x="scores", y="annotators", orient="h")
plt.title("10 Lowest MACE Competence Scores")
```

```python
sns.histplot(scoreDf["scores"])
plt.title("MACE Competence Score Histogram")
```
